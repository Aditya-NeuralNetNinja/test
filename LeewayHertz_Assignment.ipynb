## Correlation vs. Multiple Regression:
### 1) df.corr() only provides pairwise correlation coefficients, ignoring other variables.
### 2) Multiple linear regression estimates each predictors unique effect, controlling for others.

### Thus, to measure partial effects, use regression.

Using Linear Regression Over Complex Models
------------------------------------------------------------------------------

 Interpretability: Linear regression provides a direct coefficient for crime rate,
  clarifying how median home value changes per 0.01 increase, holding other variables constant.

 Transparency: Models linearity ensures straightforward understanding;
  treebased methods often obscure the global influence of individual predictors.

 Single Coefficient: Complex tree based ML Algos lack a uniform coefficient indicating change in home value per unit of crime rate.

Hence, linear regression emerges as the most succinct and effective choice for
quantifying the precise dollar impact of crime rate on median home value.

```python
from sklearn.linear_model import LinearRegression
```

## Separating input features & target label

```python
X = df.drop(columns=["Median Home Value"])
y = df["Median Home Value"]
```

## Fit linear regression on entire data


```python
model = LinearRegression()
model.fit(X,y)
```

## Display all learned coefficients for each feature

```python
coefficients = model.coef_
print("Coefficients for each column in X:")
for col_name, coef_value in zip(X.columns, coefficients):
    print(f"  {col_name}: {coef_value:.4f}")
```

## Identify the crime-rate coefficient

```python
crime_index = list(X.columns).index("crime rate")
crime_coef = model.coef_[crime_index]
print("Crime rate coefficient (per 1.0 increase):", crime_coef)
```

## Calculate effect for a 0.01 increase

```python
effect_for_0_01 = crime_coef * 0.01
print("Effect on median home value for 0.01 increase in crime rate:", effect_for_0_01, "units")
```

```python
# If 'Median Home Value' is in thousands of dollars:
dollars_effect = effect_for_0_01 * 1000
print(f"For every 0.01 increase in crime rate, median home value decreases by ${abs(dollars_effect):,.2f}")
```
